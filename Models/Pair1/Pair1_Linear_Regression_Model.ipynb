# Preprocessing of Bajaj dataset


```python
bajaj_df = bajaj_df.reset_index()
bajaj_df.head()
```

## Creating training dataset Bajaj

```python
# Cropping dataset to be in training range
X_train_bajaj = bajaj_df[bajaj_df["Date"] < "2018-12-26"]
X_train_bajaj = X_train_bajaj[X_train_bajaj["Date"] >= "2018-10-09"]

X_train_bajaj
```

```python
# Independent variables
# features = ['High', 'Low', 'Open', 'Volume']
features = ['Volume', 'Low', 'Open']
X_train_bajaj = X_train_bajaj[features]

X_train_bajaj
```

```python
len(X_train_bajaj)
```

```python
# Cropping dataset to be in training range
y_train_bajaj = bajaj_df[bajaj_df["Date"] <= "2018-12-26"]
y_train_bajaj = y_train_bajaj[y_train_bajaj["Date"] >= "2018-10-10"]

y_train_bajaj
```

```python
y_train_bajaj = y_train_bajaj['Close']

y_train_bajaj
```

```python
len(y_train_bajaj)
```

# Create and train Model for Bajaj

```python
bajaj_model = linear_model.LinearRegression()
bajaj_model = bajaj_model.fit(X_train_bajaj, y_train_bajaj)
```

```python
bajaj_model.intercept_
```

```python
bajaj_train_predictions = bajaj_model.predict(X_train_bajaj)
```

```python
bajaj_train_predictions
```

```python
print('Intercept: ', bajaj_model.intercept_)
print('Coefficients: ', bajaj_model.coef_)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_train_bajaj, bajaj_train_predictions))  
print('Mean Squared Error:', metrics.mean_squared_error(y_train_bajaj, bajaj_train_predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_train_bajaj, bajaj_train_predictions)))
print('R2 Score:', metrics.r2_score(y_train_bajaj, bajaj_train_predictions))
```

```python
stats.summary(bajaj_model, X_train_bajaj, y_train_bajaj, features)
```

## Making predictions for Bajaj

```python
# Getting data to predict in range
X_pred_bajaj = bajaj_df[bajaj_df["Date"] >= "2018-12-26"]
X_pred_bajaj = X_pred_bajaj[X_pred_bajaj["Date"] <= "2019-01-04"]

X_pred_bajaj
```

```python
# Removing columns
X_pred_bajaj = X_pred_bajaj[features]

X_pred_bajaj
```

```python
len(X_pred_bajaj)
```

```python
# Getting actual data of that period to find accuracy of predicitions
y_actual_bajaj = bajaj_df[bajaj_df["Date"] >= "2018-12-26"]
y_actual_bajaj = y_actual_bajaj[y_actual_bajaj["Date"] <= "2019-01-04"]

y_actual_bajaj
```

```python
# Extracting actual close data
y_actual_bajaj = y_actual_bajaj['Close']

y_actual_bajaj
```

```python
len(y_actual_bajaj)
```

```python
y_pred_bajaj = bajaj_model.predict(X_pred_bajaj)
```

```python
y_pred_bajaj
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_pred_bajaj, y_actual_bajaj))  
print('Mean Squared Error:', metrics.mean_squared_error(y_pred_bajaj, y_actual_bajaj))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_pred_bajaj, y_actual_bajaj)))
print('R2 Score:', metrics.r2_score(y_pred_bajaj, y_actual_bajaj))
```

```python
stats.summary(bajaj_model, X_pred_bajaj, y_actual_bajaj, features)
```
